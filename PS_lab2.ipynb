```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#zmienne dla zad1
a0=0.853333
a1=-0.518764
a2=0.129691

def f(t):
    return 1.28*t**2
    
def fa1(t):
    return a1*np.cos(np.pi*t)
    
def fa2(t):
    return a2*np.cos(2*np.pi*t)

def fourier(t):
  return a0/2 + fa1(t) + fa2(t) 
```

```python
t=np.linspace(-1,1,100)
fig, ax = plt.subplots()
ax.plot(t,f(t),t,fourier(t))
ax.grid(True)
ax.set_title('Wykresy funkcji')
plt.legend(["F. oryginalna", "T. Fouriera"])
plt.show
```

wzór ogólny na liczenie szeregu fouriera

```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
from scipy import integrate

def f(t):
    return 1.28*t**2

def ak(k):
    return (2/T) * quad(lambda t: f(t)*np.cos(2*np.pi*k*t/T), -T/2, T/2)[0]

T = 6 #okres funkcji
k = [0,1,2]
ak_value = []
for item in k:
  ak_value.append(ak(item))

print(ak_value)
k=15 #liczba wyrazów ciągu
def fourier_series(t):
    a0 = ak(0)
    sum = a0/2
    for n in range(1, k+1):
        sum += ak(n) * np.cos(2*np.pi*n*t/T)
    return sum

t=np.linspace(-3,3,1000)
fig, ax = plt.subplots()
ax.plot(t,f(t),t,fourier_series(t))
ax.grid(True)
ax.set_title('Wykresy funkcji')
plt.legend(["F. oryginalna", "T. Fouriera"])
plt.show
```
